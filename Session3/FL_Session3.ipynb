## **2.1. Caracteres Especiais e Sequências de Escape**

Ao trabalhar com ficheiros de texto é comum lidarmos com **sequências de escape** — combinações de caracteres especiais iniciadas com uma barra invertida `\`.

Estas sequências permitem representar **caracteres invisíveis ou especiais**, como quebras de linha, tabulações, aspas, entre outros.

**Tabela de Sequências de Escape em Python**

| Sequência | Significado                         | Exemplo (String)             | Resultado real                   |
|-----------|-------------------------------------|-------------------------------|----------------------------------|
| `\n`      | Nova linha (*newline*)              | `"linha 1\nlinha 2"`          | linha 1 <br> linha 2             |
| `\t`      | Tabulação horizontal                | `"palavra1\tpalavra2"`        | palavra1    palavra2             |
| `\\`      | Barra invertida literal (`\`)       | `"c:\\ficheiro\\texto.txt"`   | c:\ficheiro\texto.txt            |

````python
# Exemplo com \n (nova linha) e \t (tabulação)
print("Esta é a primeira linha.\nEsta é a segunda linha.\n")
print("Nome:\tMachado de Assis")
````

## **2.2. Manipulação de Ficheiros em Python**

Quando trabalhamos com grandes quantidades de texto é comum termos de **ler, escrever e processar ficheiros de texto**.
O python oferece uma forma simples de interagir com ficheiros através da função `open()`.

| Conceito                        | Descrição                                                                 |
|--------------------------------|---------------------------------------------------------------------------|
| `open('ficheiro.txt')`         | Abre um ficheiro em modo de leitura (por omissão).                        |
| `open('ficheiro.txt', 'w')`    | Abre um ficheiro em modo de escrita (apaga conteúdo anterior).            |
| `open('ficheiro.txt', 'a')`    | Abre um ficheiro em modo de adição (escreve no final).                    |
| `open('ficheiro.txt', 'r')`    | Abre um ficheiro em modo de leitura.                                      |
| `read()`                       | Lê todo o conteúdo do ficheiro.                                           |
| `readline()`                   | Lê uma linha do ficheiro.                                                 |
| `readlines()`                  | Lê todas as linhas e devolve uma lista.                                   |
| `write(texto)`                 | Escreve texto no ficheiro.                                                |
| `close()`                      | Fecha o ficheiro. É importante para libertar recursos.                    |
| `with open(...) as f:`         | Abre o ficheiro de forma segura (fecha automaticamente após uso).         |

**Exemplo 1: Ler um ficheiro linha a linha**

````python
x = open('text.txt', 'r', encoding='utf-8')
for i in x:
    print(i.strip())
````

````python
texto = ""

with open('text.txt', 'r', encoding='utf-8') as ficheiro:
    for linha in ficheiro:
        texto += linha
print(texto)
````

**Exemplo 2: Escrever num ficheiro**

````python
with open('novo_ficheiro.txt', 'w', encoding='utf-8') as f:  # Cuidado que apaga o que está previamente no ficheiro
    f.write("\tPrimeira linha.\n")
    f.write("Segunda linha.")
````

**Exemplo 2: Adicionar texto a um ficheiro**

````python
with open('text.txt', 'a', encoding='utf-8') as f:  # Aqui vamos acrescentar ao ficheiro existente
    f.write("\nTeste 1.")
````

````python
with open('text.txt', 'r', encoding='utf-8') as ficheiro:
    for linha in ficheiro:
        print(linha.strip())
````

**Exemplo 3: Remover apenas a linha adicionada**

````python
# Supondo que queres remover a linha "Linha extra"
with open('text.txt', 'r', encoding='utf-8') as f:
    linhas = f.readlines()
linhas
````

````python
linhas_filtradas = [linha.strip() for linha in linhas if linha.strip() != 'Teste 1.']
linhas_filtradas
````

````python
for linha in range(len(linhas_filtradas) - 1):
    linhas_filtradas[linha] += '\n'  # Adiciona uma nova linha ao final de cada linha filtrada
linhas_filtradas
````

````python
with open('text.txt', 'w', encoding='utf-8') as f:
    f.writelines(linhas_filtradas)
````

**Exemplo 4: Escrever várias linhas de uma só vez**

````python
linhas = ["Linha 1\n", "Linha 2\n", "Linha 3"]
with open('novo_ficheiro.txt', 'w', encoding='utf-8') as f:
    f.writelines(linhas)
````

````python
import re

palavras_contar = []
dict_contar = {}

with open('text.txt', 'r', encoding='utf-8') as ficheiro:
    for linha in ficheiro:
        linha_limpa = linha.strip()
        for palavra in linha_limpa.split():
            texto_limpo = re.sub(r'[^\w\s]', '', palavra.lower())
            palavras_contar.append(texto_limpo)

for palavra in palavras_contar:
    if palavra in dict_contar:
        dict_contar[palavra] += 1
    else:
        dict_contar[palavra] = 1

print(dict_contar)
````

````python
import matplotlib.pyplot as plt

plt.bar(dict_contar.keys(), dict_contar.values())
````

### **3.1.2. Lineplot**
- O lineplot (gráfico de linhas) é uma representação visual que mostra a evolução ou variação de um valor ao longo de um eixo, normalmente o tempo ou a posição.
- Neste caso, vamos utilizar um lineplot para visualizar o número de caracteres por linha no ficheiro de texto. Esta abordagem ajuda a perceber como varia o comprimento das linhas ao longo da obra, o que pode ser relevante para identificar estruturas formais, mudanças de estilo ou seções distintas.

````python
def plot_lineplot(abcissas, odenadas, titulo='Palavras mais frequentes'):
    # Criar o gráfico de linha
    plt.figure(figsize=(10, 6))
    plt.plot(abcissas, odenadas, marker='o', linestyle='-', color='blue', linewidth=2, markersize=8)

    # Personalização do gráfico
    plt.title(titulo)
    plt.xlabel('Tempo')
    plt.ylabel('Frequencia') 
    plt.grid(True)
    plt.legend()
    plt.tight_layout()

    # Mostrar
    plt.show()
````

````python
alinea_count = {}
with open('text.txt', 'r', encoding='utf-8') as file:
    for number, linha in enumerate(file):
        print(f"Linha {number}: {len(linha.strip())}")
        alinea_count[number] = len(linha.strip())
plot_lineplot(alinea_count.keys(), alinea_count.values(), titulo='Contagem de Alíneas no Ficheiro')
````

### **3.1.3. Pie Chart**
- O pie chart (gráfico circular) é uma representação gráfica usada para mostrar a proporção de diferentes categorias dentro de um todo.
- Vamos utilizá-lo para visualizar a distribuição das vogais no ficheiro de texto, permitindo perceber quais são mais frequentes. Esta análise pode ser útil em estudos linguísticos e fonológicos dentro do contexto da filologia digital.

````python
def plot_piechart(labels, valores, titulo='Distribuição de categorias'):
    """Cria um gráfico circular a partir de categorias e respetivos valores."""
    plt.figure(figsize=(8, 8))
    plt.pie(valores, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

    # Personalização
    plt.title(titulo)
    plt.axis('equal')  # Garante que o círculo fique redondo
    plt.tight_layout()

    # Mostrar
    plt.show()
````

````python
vogal_count = {"a": 0, "e": 0, "i": 0, "o": 0, "u": 0}
with open('text.txt', 'r', encoding='utf-8') as file:
    for linha in file:
        for char in linha.lower():
            if char in vogal_count:
                vogal_count[char] += 1
plot_piechart(vogal_count.keys(), vogal_count.values(), titulo='Distribuição de Vogais no Ficheiro')
````

<a class='anchor' id='3.1.4.'></a>
### **3.1.4. Scatter Plot**
- O scatter plot (gráfico de dispersão) é uma visualização que mostra a relação entre duas variáveis numéricas.
- Neste contexto, será utilizado para representar a relação entre o número de palavras e o número de caracteres por linha do ficheiro analisado.
- Este tipo de gráfico é útil para perceber, por exemplo, se as linhas mais longas (em caracteres) também tendem a ter mais palavras, ou se existem linhas curtas mas densas em termos linguísticos.

````python
def plot_scatter(x, y, titulo='Relação entre variáveis', xlabel='Variável X', ylabel='Variável Y'):
    """Cria um gráfico de dispersão (scatter plot) a partir de duas listas de valores."""
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='teal', alpha=0.7, edgecolors='black')

    # Personalização
    plt.title(titulo)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.tight_layout()

    # Mostrar
    plt.show()
````

````python
comp_freq = {}

with open('text.txt', 'r', encoding='utf-8') as file:
    for linha in file:
        palavras = linha.strip().split()
        palavras = [palavra.strip(",.;") for palavra in palavras]
        for palavra in palavras:
            if len(palavra) in comp_freq.keys():
                comp_freq[len(palavra)] += 1
            else:
                comp_freq[len(palavra)] = 1


plot_scatter(comp_freq.keys(), comp_freq.values(),
             titulo='Comprimento das palavras vs. frequência',
             xlabel='Comprimento da palavra',
             ylabel='Frequência')
````

### **3.2.1. Barplot**
- Abaixo apresenta-se uma função simples para gerar um gráfico de barras (barplot). Iremos visualizar a frequência de palavras contidas no ficheiro, seguindo-se a representação das 10 palavras mais frequentes.

````python
# Função para barplot com Seaborn + numeração
def plot_barplot_sns(x, y, titulo='Frequência de Palavras', horizontal=False):
    plt.figure(figsize=(12, 7))
    data = list(zip(x, y))
    data = sorted(data, key=lambda item: item[1], reverse=True)
    palavras, frequencias = zip(*data)

    ax = None

    if horizontal:
        ax = sns.barplot(x=frequencias, y=palavras, palette="viridis")
        plt.xlabel('Frequência')
        plt.ylabel('Palavras')

        # Adicionar os valores no final das barras
        for i, v in enumerate(frequencias):
            ax.text(v + 0.1, i, str(v), color='black', va='center', fontweight='bold')
    else:
        ax = sns.barplot(x=palavras, y=frequencias, palette="viridis")
        plt.xlabel('Palavras')
        plt.ylabel('Frequência')
        plt.xticks(rotation=45)

        # Adicionar os valores no topo das barras
        for i, v in enumerate(frequencias):
            ax.text(i, v + 0.1, str(v), color='black', ha='center', fontweight='bold')

    plt.title(titulo, fontsize=16)
    plt.tight_layout()
    plt.show()

plot_barplot_sns(high_contar.keys(), high_contar.values(), titulo='Top10 Palavras Mais Frequentes', horizontal=True)
````

### **3.2.2. Lineplot**
- O lineplot (gráfico de linhas) é uma representação visual que mostra a evolução ou variação de um valor ao longo de um eixo, normalmente o tempo ou a posição.
- Neste caso, vamos utilizar um lineplot para visualizar o número de caracteres por linha no ficheiro de texto. Esta abordagem ajuda a perceber como varia o comprimento das linhas ao longo da obra, o que pode ser relevante para identificar estruturas formais, mudanças de estilo ou seções distintas.

````python
def plot_lineplot_sns(x, y, titulo='Número de Caracteres por Linha'):

    plt.figure(figsize=(14, 7))
    ax = sns.lineplot(
        x=list(x), y=list(y), 
        marker='o', 
        markersize=10, 
        linewidth=3, 
        color=sns.color_palette("deep")[0],
        markerfacecolor='black', 
        markeredgewidth=2
    )
    
    # Adicionar valores em cima dos pontos
    for xi, yi in zip(x, y):
        ax.text(xi, yi + max(y)*0.02, str(yi), ha='center', va='bottom', fontsize=12, weight='bold')
    
    ax.set_title(titulo, fontsize=20, weight='bold')
    ax.set_xlabel('Número da Linha', fontsize=14)
    ax.set_ylabel('Número de Caracteres', fontsize=14)
    ax.grid(True, which='major', linestyle='--', linewidth=0.8, alpha=0.7)
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    plt.show()

# Exemplo de chamada
plot_lineplot_sns(alinea_count.keys(), alinea_count.values(), titulo='Contagem de Alíneas no Ficheiro')
````

### **3.2.3. Pie Chart**
- O pie chart (gráfico circular) é uma representação gráfica usada para mostrar a proporção de diferentes categorias dentro de um todo.
- Vamos utilizá-lo para visualizar a distribuição das vogais no ficheiro de texto, permitindo perceber quais são mais frequentes. Esta análise pode ser útil em estudos linguísticos e fonológicos dentro do contexto da filologia digital.

````python
def plot_piechart_beautiful(labels, valores, titulo='Distribuição de categorias'):
    plt.figure(figsize=(8, 8))
    
    # Paleta de cores Seaborn
    cores = sns.color_palette('pastel')[0:len(labels)]
    
    # Explode para destacar a maior fatia (opcional)
    explode = [0.1 if v == max(valores) else 0 for v in valores]
    
    wedges, texts, autotexts = plt.pie(
        valores,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        colors=cores,
        explode=explode,
        shadow=True,
        textprops={'fontsize': 14, 'weight': 'bold', 'color': 'navy'}
    )
    
    plt.title(titulo, fontsize=18, weight='bold')
    plt.axis('equal')  # Círculo perfeito
    plt.tight_layout()
    plt.show()

plot_piechart_beautiful(vogal_count.keys(), vogal_count.values(), titulo='Distribuição de Vogais no Ficheiro')
````

- Neste contexto, será utilizado para representar a relação entre o número de palavras e o número de caracteres por linha do ficheiro analisado.
- Este tipo de gráfico é útil para perceber, por exemplo, se as linhas mais longas (em caracteres) também tendem a ter mais palavras, ou se existem linhas curtas mas densas em termos linguísticos.

````python
def plot_scatter_sns(x, y, titulo='Relação entre Comprimento de Palavra e Frequência', xlabel='Comprimento da Palavra', ylabel='Frequência'):
    plt.figure(figsize=(14, 8))
    ax = sns.scatterplot(
        x=list(x),
        y=list(y),
        s=150,                        # tamanho dos pontos
        color=sns.color_palette("dark")[2],
        edgecolor='black',
        linewidth=1,
        alpha=0.85
    )

    # Anotar os valores em cada ponto (frequência)
    for xi, yi in zip(x, y):
        ax.text(xi, yi + max(y)*0.02, str(yi), 
                ha='center', va='bottom', 
                fontsize=12, weight='bold', color='navy')

    ax.set_title(titulo, fontsize=20, weight='bold')
    ax.set_xlabel(xlabel, fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)

    # Melhorar grelha
    ax.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    plt.show()

# Chamada exemplo
plot_scatter_sns(comp_freq.keys(), comp_freq.values(),
                 titulo='Comprimento das palavras vs. frequência',
                 xlabel='Comprimento da palavra',
                 ylabel='Frequência')
````

### **3.3.2. Gráfico de Rede de Coocorrência**
-  Gráfico de Rede de Coocorrência é uma visualização que representa a relação entre palavras com base na sua proximidade num texto. Sempre que duas palavras aparecem próximas uma da outra (dentro de uma janela definida), é criada uma ligação (ou aresta) entre elas.
- Neste tipo de grafo, as palavras funcionam como nós (vértices), e a força da ligação entre elas é representada pela espessura das arestas, que depende da frequência com que essas palavras ocorrem juntas.
- Esta técnica é particularmente útil para explorar estruturas semânticas e padrões de coocorrência lexical, ajudando a identificar grupos de palavras que aparecem frequentemente em conjunto. Pode ser aplicada em estudos linguísticos, análise de tópicos, estilometria ou em sistemas de recomendação textual.
- Nesta implementação, o grafo é construído com base nas 30 coocorrências mais frequentes, considerando uma janela de ``n`` palavras (antes e depois). A visualização destaca as ligações mais significativas, permitindo uma análise visual clara das associações léxicas dominantes no texto analisado.

````python
!pip install networkx
````

````python
import networkx as nx
from collections import Counter
import re

def plot_coocurrence_graph(filepath='text.txt', window_size=3, top_n=30, titulo='Rede de Coocorrência de Palavras'):
    """Cria um gráfico de rede (grafo) com base na coocorrência de palavras num ficheiro de texto."""

    # Ler e pré-processar o texto
    with open(filepath, 'r', encoding='utf-8') as f:
        texto = f.read().lower()
        texto = re.sub(r'[^\w\s]', '', texto)  # Remove pontuação
        palavras = texto.split()

    # Calcular coocorrências dentro da janela definida
    coocorrencias = Counter()
    for i, palavra in enumerate(palavras):
        janela = palavras[max(i - window_size, 0): i] + palavras[i+1: i+1 + window_size]
        for vizinha in janela:
            par = tuple(sorted((palavra, vizinha)))
            coocorrencias[par] += 1

    # Selecionar os pares mais frequentes
    top_pares = dict(coocorrencias.most_common(top_n))

    # Criar grafo
    G = nx.Graph()
    for (w1, w2), peso in top_pares.items():
        G.add_edge(w1, w2, weight=peso)

    # Visualizar grafo
    plt.figure(figsize=(14, 9))
    pos = nx.spring_layout(G, k=0.9, seed=42)  # Layout com espaçamento mais claro
    edges = G.edges(data=True)
    pesos = [dados['weight'] for _, _, dados in edges]

    # Desenhar nós e arestas com gradientes
    nx.draw_networkx_nodes(G, pos, node_size=800, node_color='lightsteelblue', edgecolors='black')
    nx.draw_networkx_edges(G, pos, width=[p * 0.4 for p in pesos], alpha=0.5, edge_color='grey')
    nx.draw_networkx_labels(G, pos, font_size=11, font_weight='bold', font_family='sans-serif')

    plt.title(titulo, fontsize=18, weight='bold')
    plt.axis('off')
    plt.tight_layout()
    plt.show()

plot_coocurrence_graph(filepath='text.txt', window_size=3, top_n=35)
````

````python
import requests

url = "https://www.gutenberg.org/ebooks/55752.txt.utf-8"

try:
    resp = requests.get(url, timeout=15)  # timeout de 15 segundos
    resp.encoding = 'utf-8'

    if resp.status_code == 200:
        texto = resp.text
        with open("dom_casmurro.txt", "w", encoding="utf-8") as f:
            f.write(texto)
        print("Ficheiro 'dom_casmurro.txt' criado com sucesso!")
    else:
        print("Erro ao descarregar:", resp.status_code)

except requests.exceptions.ConnectTimeout:
    print("Ligação expirada: o servidor não respondeu a tempo.")
except requests.exceptions.RequestException as e:
    print(f"Erro ao tentar aceder ao recurso: {e}")
````

````python
def limpar_e_atualizar_ficheiro(filepath):
    with open(filepath, encoding='utf-8') as f:
        linhas = f.readlines()
    
    start_idx = 0
    end_idx = len(linhas)
    
    # Encontrar início do texto real
    for i, linha in enumerate(linhas):
        if 'Do titulo.' in linha:
            start_idx = i + 1
            break
    
    # Encontrar fim do texto real
    for i, linha in enumerate(linhas):
        if 'FIM' in linha:
            end_idx = i
            break
    
    # Extrair texto principal e limpar espaços e linhas vazias
    texto_limpo = [linha.strip() for linha in linhas[start_idx:end_idx] if linha.strip()]
    
    # Sobrescrever o ficheiro com o texto limpo
    with open(filepath, 'w', encoding='utf-8') as f:
        f.write("\n".join(texto_limpo))

# Chama a função para atualizar o ficheiro
limpar_e_atualizar_ficheiro("dom_casmurro.txt")

print("Ficheiro 'dom_casmurro.txt' atualizado com texto limpo!")
````
