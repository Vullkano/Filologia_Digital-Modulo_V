# Módulo V: Do Pergaminho ao Computador  
## Editar Manuscritos na Era Digital

## Tópicos Abordados

### 1. Motivação para Usar uma Linguagem de Programação

**Objetivo:**  
Apresentar as vantagens e a importância de aprender a programar, especialmente no contexto da edição digital de manuscritos.

**Ferramentas:**  
Jupyter Notebook, Google Colab, VSCode, terminal Python, Anaconda.

**Conteúdo:**  
- Introdução à sintaxe e ao vocabulário básico da linguagem Python.  
- Demonstrações simples com foco na aplicabilidade prática.

**Notebook:**  
[Session 1](Session1/FL_Session1.ipynb)

### 2. Fundamentos de Python

**Objetivo:**  
Explorar as estruturas fundamentais da linguagem Python e a sua utilidade na manipulação de texto.

**Conteúdo:**  
- Estruturas de controlo: ciclos `for` e `while`.  
- Estruturas de dados: listas, tuplos, dicionários, conjuntos.  
- Tipos de dados e operadores.  
- Expressões regulares para extração e filtragem de texto.  
- Exemplos práticos:  
  - Filtragem de listas com base em sufixos.  
  - Cruzamento de listas (e.g. filtragem com base noutra lista).

**Notebook:**  
[Session 2](Session2/FL_Session2.ipynb)

### 3. Introdução ao Pandas e Visualização de Dados

**Objetivo:**  
Ensinar a manipulação de dados estruturados e a criação de visualizações informativas.

**Conteúdo:**  
- Criação e manipulação de DataFrames com Pandas.  
- Operações em tabelas: filtragem, ordenação, agrupamento.  
- Leitura e escrita de ficheiros (CSV, Excel, XML).  
- Visualização de dados com Matplotlib e Seaborn:  
  - Gráficos de barras, histogramas, boxplots, etc.

**Notebook:**  
[Session 3](Session3/FL_Session3.ipynb)

### 4. Fundamentos de Machine Learning e NLP

**Objetivo:**  
Introduzir os conceitos fundamentais de Machine Learning e Processamento de Linguagem Natural (NLP).

**Conteúdo:**  
- Aprendizagem supervisionada vs. não supervisionada.  
- Métricas de avaliação: accuracy, precision, recall, F1-score.  
- Introdução a bibliotecas: NLTK, spaCy, TextBlob, Scikit-learn.  
- Exemplo prático:  
  - Treino de um classificador de sentimentos com dados textuais simples.

### 5. Continuação e Consolidação

**Objetivo:**  
Reforçar os conceitos aprendidos através de práticas adicionais e esclarecimento de dúvidas.

**Conteúdo:**  
- Sessões de resolução de exercícios e mini-projetos.  
- Revisão dos tópicos anteriores.  
- Aprofundamento consoante o interesse e ritmo dos participantes.
